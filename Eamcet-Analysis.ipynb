> # " DATA CLEANING "


```python
Eamcet.isna()
```

```python
#See if there are any null values 
#Here we can see that some null values are present
Eamcet.isnull().sum()
```

```python
Eamcet['OC BOYS'] = Eamcet['OC BOYS'].replace(np.nan,"12")
Eamcet['BC_A\nBOYS'] = Eamcet['BC_A\nBOYS'].replace(np.nan,"356")
Eamcet['BC_B\nBOYS'] = Eamcet['BC_B\nBOYS'].replace(np.nan,"36")
Eamcet['BC_C\nBOYS'] = Eamcet['BC_C\nBOYS'].replace(np.nan,"56")
Eamcet['BC_D\nBOYS'] = Eamcet['BC_D\nBOYS'].replace(np.nan,"43")
Eamcet['BC_E\nBOYS'] = Eamcet['BC_E\nBOYS'].replace(np.nan,"26")
Eamcet['SC\nBOYS'] = Eamcet['SC\nBOYS'].replace(np.nan,"38")
Eamcet['ST\nBOYS'] = Eamcet['ST\nBOYS'].replace(np.nan,"40")
Eamcet['EWS GEN OU'] = Eamcet['EWS GEN OU'].replace(np.nan,"100")
```

```python
#Now you can see that all null values are cleared 
Eamcet.isnull().sum()
```

```python
#Here we have removed the INST CODE from the dataset
Eamcet= Eamcet.drop(['INST\nCODE'],axis=1)
Eamcet.head()
```

```python
#Here we have removed the TYPE from the dataset
Eamcet= Eamcet.drop(['TYPE'],axis=1)
Eamcet.head()
```

> # " DATA MANIPULATION "

```python
#Here we are converting the each column from string to int 
#And merging all the OC Section girls and boys and creating the column name as OC students
Eamcet['OC_students'] = Eamcet['OC BOYS'].astype(int) + Eamcet['OC GIRLS'].astype(int)
Eamcet['OC_students']
```

```python
#Here we are converting the each column from string to int
#And merging all the BC Section girls and boys and creating the column name as BC students
Eamcet['BC_students'] = Eamcet['BC_A\nBOYS'].astype(int) + Eamcet['BC_A\nGIRLS'].astype(int)+Eamcet['BC_B\nBOYS'].astype(int)+Eamcet['BC_B\nGIRLS'].astype(int)+Eamcet['BC_C\nBOYS'].astype(int)+Eamcet['BC_C\nGIRLS'].astype(int)+Eamcet['BC_D\nBOYS'].astype(int)+Eamcet['BC_D\nGIRLS'].astype(int)+Eamcet['BC_E\nBOYS'].astype(int)+Eamcet['BC_E\nGIRLS'].astype(int)
print(Eamcet['BC_students'])
```

```python
#Here we are converting the each column from string to int 
#And merging all the SC Section girls and boys and creating the column name as SC students
Eamcet['SC_students'] = Eamcet['SC\nBOYS'].astype(int) + Eamcet['SC\nGIRLS'].astype(int)
Eamcet['SC_students']
```

```python
#Here we are converting the each column from string to int
#And merging all the ST Section girls and boys and creating the column name as ST students
Eamcet['ST_students'] = Eamcet['ST\nBOYS'].astype(int) + Eamcet['ST\nGIRLS'].astype(int)
Eamcet['ST_students']
```

```python
#Showing BC OC SC ST categories by donut chart
labels=["All_BC_students","All_OC_students","All_SC_students","All_ST_students"]
values= [All_BC_students.values[:10].sum(),
        All_OC_students.values[:10].sum(),
        All_SC_students.values[:10].sum(),
        All_ST_students.values[:10].sum()]
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.4)])
fig.update_layout(
    title_text='All Category Students',
    title_x=0.5,  # Center the title
    margin=dict(t=50),  # Set top margin to create space for the title
)
fig.update_traces(textinfo='label+percent', hovertemplate='%{label}: %{customdata[1]}<br>(%{percent})')

fig.show()
```

```python
custom_colors = ['#5f9ea0']
fig = px.histogram(Eamcet, 
                   x='TUITION FEE',
                   color_discrete_sequence=custom_colors,
                   title='Histogram for TUITION FEE',
                   labels={'TUITION FEE': 'Tuition Fee'},
                   height=400
                   )

fig.show()
```

```python
College = Eamcet['YEAR OF\nESTB'].value_counts().reset_index()
College.columns = ['YEAR OF\nESTB', 'Count']
bv=College.head(10)
plt = px.pie(bv, names='YEAR OF\nESTB', values='Count', title='Institutes that are Built From 1998 to 2008')
plt.update_traces(textinfo='label+percent', hovertemplate='%{label}: %{value} (%{percent})')

plt.show()

```

```python
#plotting the values of only just CSE and ECE by using Box plot
branches = ['CSE', 'ECE']
filtered_df = Eamcet[Eamcet['BRANCH'].isin(branches)]
colors = ['#de3163', '#ff7f0e']
fig = px.box(filtered_df, 
             x='BRANCH', 
             y='TUITION FEE',
             points='all',
             color_discrete_sequence=colors,
             title='Box Plot for CSE and ECE Tuition Fees',
             labels={'BRANCH': 'Branch', 'TUITION FEE': 'Tuition Fee'},
             )

fig.show()
```

```python
top_colleges = Eamcet['INSTITUTE NAME'].value_counts()[:15]
top_colleges_names = list(top_colleges.index)
top_colleges_names
```
